```python
import numpy as np
import pandas as pd

#데이터셋 불러오기
anime=pd.read_csv("data/anime.csv")
rating=pd.read_csv("data/rating.csv")
```

```python
#데이터셋 살펴보기
display(anime.info())
display(anime.describe())
display(anime.head())
```

```python
#데이터셋 살펴보기
display(rating.info())
display(rating.describe())
display(rating.head())

anime
```

```python
#특수문자가 섞인 name column에서 특수문자 지우기
anime.name=anime.name.str.replace('&quot;','',regex=True)
anime.name=anime.name.str.replace('&#039;','',regex=True)
anime
```

```python
#결측치 제거
anime=anime.dropna(axis=0)
rating=rating.dropna(axis=0)
display(anime.info())
display(rating.info())
```

```python
rating=rating[rating.rating!=-1]
rating.reset_index(drop=True, inplace=True)
rating
```

```python
#애니메이션 id의 list 만들기, 중복 값 제거
anime_id_list=rating["anime_id"].unique()
#사용자 평가가 없는 애니메이션 필터링
anime=anime[anime["anime_id"].isin(anime_id_list)]
anime.reset_index(drop=True, inplace=True)
```

```python
#각 애니메이션이 가질 수 있는 타입 값 확인하기
type_list=anime["type"].unique()
print(type_list)
```

```python
#anime에 새로운 column(type_list)을 만듦.
anime[type_list]=np.NaN
#anime의 타입별로 True/False를 넣어줌.
for t in type_list:
    anime[t]=np.where(anime["type"]==t, True,False)
anime=anime.drop(columns=["type"])
```

```python
#모든 행의 장르 값을 담을 문자열 변수 생성
genre_str=""
#모든 행을 돌며 장르의 값을 더하고, 행마다 구분자로 ,를 넣어줌.
for i in range(len(anime['genre'])):
    genre_str+=str(anime['genre'][i])
    genre_str+=","
```

```python
#공백제거, 장르 문자열을 구분자 ','를 기준으로 자름.
genre_str=genre_str.replace(" ","")
genre_temp=genre_str.split(",")
#중복 제거
genre_list=set(genre_temp)
genre_list.remove('')
genre_list=list(genre_list)
print(genre_list)
```

```python
#anime에 새로운 column(genre_list)을 만듦.
anime[genre_list]=np.NaN
#anime의 genre별로 column을 생성하고, 장르 값을 가진 경우 True, 아닌 경우 false로 만듦.
for g in genre_list:
    anime[g]=np.where(anime.genre.str.find(g)!=-1,True,False)

anime=anime.drop(columns=['genre'])
```

```python
#rating dataframe과 merge를 위해 index를 바꿔줌
anime=anime.set_index("anime_id")
display(anime)
```

```python
#rating 데이터에 anime의 name 행을 추가
rating=pd.merge(rating, anime["name"],on="anime_id")
rating
```

```python
#애니메이션 제목이 column이 되도록 table pivot
rating = rating.pivot_table(values="rating",index="user_id", columns="name")
rating
```

```python
#결측치 처리하기
rating=rating.fillna(0)
rating
```

```python
#SVD 모델에 넣기 전 행과 열을 반전시킴
rating_matrix=rating.values.T
#anime_num, user_num이 각각 9888, 69600 값을 가짐. 
print(rating_matrix.shape)
```

```python
#TruncatedSVD 모델을 사용하기 위해 패키지 불러옴
from sklearn.decomposition import TruncatedSVD
#12개의 주요 component로 TruncatedSVD 변환
SVD=TruncatedSVD(n_components=12)
#fit_transform()을 호출해 원본 데이터를 12개의 주요 컴포넌트로 차원을 축소해 변환
matrix=SVD.fit_transform(rating_matrix)
matrix.shape
```

```python
#TruncatedSVD를 사용해 뽑아낸 정보를 피어슨 상관계수를 가져와 계산하여, 
#가장 비슷한 애니메이션을 찾을 수 있도록 해줌
corr=np.corrcoef(matrix)
corr.shape
```

```python
print(corr)
```

```python
#anime 이름들을 담은 변수 생성
#anime 값은 index로 pivot 했으므로, column값을 가져옴
anime_name_list=list(rating.columns)
anime_name_list
```

```python
#사용자에게 좋아하는 animation를 물어보기
user_selected_title=input("Input Animation Title :")
#anime_name_list에서 selected num을 찾은 후, 해당되는 상관계수 값들을 찾아서 similarity 변수에 담아둠.
selected_index=anime_name_list.index(user_selected_title)
similarity=corr[selected_index]

#인덱스를 돌며 similarity가 높은 애니메이션들을 list에 넣음
index=0
index_list={}
for val in similarity:
    if val>=0.9:
        index_list[val]=anime_name_list[index]
    index+=1
    
#similarity가 높은 순으로 sorting하고 사용자가 선택한 애니메이션은 빼줌. 20개만 출력
anime_name=[]
limit=20
index=0
for key, val in sorted(index_list.items(), reverse=True):
    if limit<index:
        break
        
    if index==0:
        index+=1
        continue;
    
    anime_name.append(val)
    index+=1
#user input, recommendation result
print("\n\n*user_input*\n\n")
display(anime[anime["name"]==user_selected_title])
print("\n\n*recommendation*")
display(anime[anime["name"].isin(anime_name)])
```
